```python
from random import random
import torch
import torch.nn as nn 
```

```python
w = [random()*100 for _ in range(5)]
w
```

```python
DataX = []
DataXIncomplete = []
DataY = []
for _ in range(10):
    x = [random() for _ in range(5)]
    DataX.append(x)
    DataXIncomplete.append(x[2])
    DataY.append(sum([x[i]*w[i] for i in range(5)]))
print(DataX[:5])
print(DataY[:5])
```

```python
from torch.utils.data import DataLoader, TensorDataset

tensor_x = torch.tensor(DataX, dtype=torch.float32)
tensor_y = torch.tensor(DataY, dtype=torch.float32)

dataset = TensorDataset(tensor_x, tensor_y)

dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
class model(nn.Module):
    def __init__(self):
        super(model, self).__init__()
        self.fc = nn.Linear(5, 1)
    def forward(self, x):
        return self.fc(x)
```

```python
model = model()
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=1, factor=0.2)
for epoch in range(1000):
    for i, (x, y) in enumerate(dataloader):
        optimizer.zero_grad()
        y_pred = model(x)
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()
        #scheduler.step(loss.item())
    print(f'Epoch {epoch} Loss {loss.item()}')
```

```python
with torch.no_grad():
    total_loss = 0
    for x, y in dataloader:
        y_pred = model(x)
        loss = criterion(y_pred, y)
        total_loss += loss.item()
    average_loss = total_loss / len(dataloader)
    print(f'Average Loss: {average_loss}')
    accuracy = 100 - (average_loss / sum(DataY) * 100)
    print(f'Accuracy: {accuracy} %')
```
